```python
#Синий круг
import cv2
import numpy as np

cap = cv2.VideoCapture('video.webm')
tap = []
i=0
while True:
    _, frame = cap.read(0)

    col, rows, ch = frame.shape

    pts1 = np.float32([[340, 215], [642, 210], [115, 500], [877, 480]])
    pts2 = np.float32([[0, 0], [650, 0], [0, 750], [650, 750]])

    matrix = cv2.getPerspectiveTransform(pts1, pts2)
    webcam = cv2.warpPerspective(frame , matrix, (660, 760))

    rob_hsv = cv2.cvtColor(webcam, cv2.COLOR_BGR2HSV) #Преобразуем в HSV
    rob_color_low = (63,75,63) #Данный цвет это темный ненасыщенный синий
    rob_color_high = (255,255,255) #Данный цвет это светлый насыщенный синий
    mask = cv2.inRange(rob_hsv, rob_color_low, rob_color_high)

    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    counturs = sorted(contours, key=cv2.contourArea, reverse=True)

    if not counturs:
        print("Робот потерян")

    for contour in counturs:
        tap.append(contour[0][0])
        if (((tap[i-1][1]-contour[0][0][1]>50) or (tap[i-1][0]-contour[0][0][0])>50) and ((contour[0][0][1]<60) and (250<contour[0][0][0]<270))):
            print("Робот потерян")
            tap.pop()
            tap.append(tap[i-1])
            cv2.circle(webcam, contour[0][0], 7, (0,0,255))
            cv2.putText(webcam, "wrong", (contour[0][0][0] - 20, contour[0][0][1] + 20), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (0, 255, 0), 2)
        else:
            cv2.drawContours(webcam, counturs[0], -1, (255, 0, 255), 3)
        
            print("Координаты робота "+str('%.3f' %(tap[i][0]*450/660))+" "+str('%.3f' %(tap[i][1]*750/760)))
        i+=1
    cv2.imshow("Counturs", webcam)  # рисует окно с конурамиq
    key = cv2.waitKey(1)
    if cv2.waitKey(10) & 0xFF == ord('q'):
            break
cap.release()
cv2.destroyAllWindows()
```

```python
#Белый лист
import cv2
import numpy as np

cap = cv2.VideoCapture('video.webm')
while True:
    _, frame = cap.read(0)

    col, rows, ch = frame.shape

    pts1 = np.float32([[340, 215], [642, 210], [115, 500], [877, 480]])
    pts2 = np.float32([[0, 0], [650, 0], [0, 750], [650, 750]])

    matrix = cv2.getPerspectiveTransform(pts1, pts2)
    webcam = cv2.warpPerspective(frame , matrix, (660, 760))
    
    low_white = (205,0,0)
    high_white = (255,255,255)
    mask = cv2.inRange(webcam, low_white, high_white)
    
    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    counturs = sorted(contours, key=cv2.contourArea, reverse=True)
    if not counturs:
        print("Робот потерян")
    for i in counturs:
        cv2.drawContours(webcam, [i], -1, (0, 255, 0), 2)
        c = max(contours, key = cv2.contourArea)
        M = cv2.moments(c)
        if M['m00'] != 0:
            cx = int(M['m10']/M['m00'])
            cy = int(M['m01']/M['m00'])
            cv2.circle(webcam, (cx, cy), 7, (255, 0, 0), -1)
            cv2.putText(webcam, "center", (cx - 20, cy - 20), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (0, 0, 0), 2)
        print("Координаты робота "+str('%.3f' %(cx*450/660))+" "+str('%.3f' %(cy*750/760)))
    cv2.imshow("Counturs", webcam)  # рисует окно с конурами
    key = cv2.waitKey(1)
    if cv2.waitKey(10) & 0xFF == ord('q'):
            break
cap.release()
cv2.destroyAllWindows()
```
